## Artificial Neuron in context

* Linear Regression: Weighted Sum of inputs
* Logistic Regression: Logistic Function(sigmoid) applied on Weighted Sum of inputs
* Artificial Neuron: Non Linear Function (Activation function)(sigmoid, tanh, relu etc) applied on Weighted Sum of inputs






## Non Linear Functions Recap

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def relu(x):
    return np.maximum(0,x)

def leaky_relu(x):
    def _leaky_relu(x):
        return x if x>0 else x*0.1
    return np.vectorize(_leaky_relu)(x)

def draw(fs):   
    X = np.linspace(-5, 5)
    fig, axes = plt.subplots(1, len(fs),  figsize=(20, 5)) #no more than 5 funcs
    for i,f in enumerate(fs):
        axes[i].axhline(c='grey')
        axes[i].axvline(c='grey')
        axes[i].set_title(f.__name__)
        axes[i].plot(X, f(X))
    
    # To plot them together
    # fig2 = plt.figure()
    # ax = fig2.add_subplot(1,1,1)
    # for f in fs:
    #     ax.plot(X, f(X))
    
draw([sigmoid, tanh,  relu, leaky_relu])
```

```python
# Solution:

import numpy as np
from sklearn.metrics import mean_squared_error

# Missing information: Weights and the Activation function

# Assuming some weight
weights = np.array([1.0, -1.0])
# Assuming activation function as relu
activation_function = relu

#Second input batch 
X = np.array([
    [1, 1],
    [2, 0],
    [0, 2]
])

Y = np.array([5, 7, 3])

#Predictions
z = np.dot(X, weights) # Weighted sum of inputs
Y_hat = activation_function(z) # activation_function applied on weighted sum of inputs
print(Y_hat)

#Prediction Error
J = mean_squared_error(Y, Y_hat) 
print(J)
```
